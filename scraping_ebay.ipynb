This script executes code to scrape ebay.com. It is necessary to input only the search terms and the number of pages and after save a CSV file with the name 'scraping_ebay.csv.


```python
# @title 1. ‚ú® Installing dependences.

!pip3 install beatifulsoup4 &> /dev/null
!pip3 install pandas &> /dev/null
!pip install aiohttp &> /dev/null
!pip install asyncio &> /dev/null

import requests
import pandas as pd
import aiohttp
import asyncio
import time

from os import link
from bs4 import BeautifulSoup

headers = {
    "User-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36"}
```

```python
# @title 2. ‚úç Run the script to scrape ebay.com.

search = "tshirt" #@param {type:"string"}
no_of_pages = 10 #@param {type:"integer"}

web_posts=[]

def scraping_ebay(search,no_of_pages):

    for page in range(no_of_pages):

            #url = 'https://www.ebay.com/sch/i.html?_from=R40&_nkw=' + search + '&_sacat=0&_pgn=' + str(page)
            url = 'https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2380057.m570.l1313&_nkw=' + search + '&_sacat=' + str(page)
            response = requests.get(url, headers=headers)
            html = response.text
            soup = BeautifulSoup(html,"html")

            # Outer Most Entry Point of HTML:
            outer_most_point=soup.find('div',attrs={'id': 'srp-river-results'})

            # "UL" lists where the data are stored:
            for i in outer_most_point.find_all('li',{'class':'s-item s-item__pl-on-bottom'}):

                # Item Title:
                item_name = i.find('div',{'class':'s-item__title'})
                if item_name != None:
                    item = item_name.find('span').text
                else:
                    item = 'No found name'

                # Item Company:
                item_price = i.find('span', {'class':'s-item__price'})
                if item_price != None:
                    price = item_price.text
                else:
                    price = 'No found price'

                # Item Country:
                item_country = i.find('span', {'class':'s-item__location s-item__itemLocation'})
                if item_country != None:
                    country = item_country.text
                else:
                    country = 'No found country'

                # Item Link:
                link = i.find('div', {'class': 's-item__info clearfix'})
                if link is not None:
                    link_a = link.find('a', {'class': 's-item__link'})
                    if link_a is not None and 'href' in link_a.attrs:
                        links = link_a['href']
                    else:
                        links = 'No found link'
                else:
                    links = 'No found link'

                # Put everything together in a list of lists
                web_posts.append([item,price,links])#,country])

    return web_posts

names=['Item','Price','links']#,'Country']

start = time.time()
scraping_df=pd.DataFrame(scraping_ebay(search,no_of_pages),columns=names)

scraping_df.to_csv('scraping_ebay.csv', index=False)

end = time.time()
print('Task completed in: ',end - start,' seconds')

web_df
```
